A basic POST request


```python
import requests
from pprint import pprint
```

```python
r_post = requests.post('https://en.wikipedia.org/w/index.php', 
                       data = {'search':'washington dc'})
```

```python
r_post.status_code
```

```python
type(r_post)
```

```python
requests.models.Response
```

```python
pprint(r_post.text)
```

```python
with open('washingtondc.html', 'wb') as f:
    for chunk in r_post.iter_content(chunk_size=10000):
        f.write(chunk)
```
